```python
import gym
import numpy as np

# Initialize the environment
env = gym.make("CartPole-v1", render_mode="human")  # Use "human" for visualization

# Number of episodes to run
num_episodes = 10

for episode in range(num_episodes):
    state = env.reset()
    state = state[0] if isinstance(state, tuple) else state  # Handle Gym versions

    done = False
    total_reward = 0

    while not done:
        env.render()  # Render the environment (optional)

        action = env.action_space.sample()  # Select a random action (0 or 1)

        # Step through environment
        step_result = env.step(action)
        if len(step_result) == 4:
            next_state, reward, done, info = step_result  # Old Gym versions
        else:
            next_state, reward, done, truncated, info = step_result  # New Gym versions
            done = done or truncated

        # Extract the pole angle (index 2 in state)
        pole_angle_rad = next_state[2]
        pole_angle_deg = np.degrees(pole_angle_rad)  # Convert to degrees

        # Print the angle at each step
        print(f"Episode {episode + 1} - Step: {total_reward + 1}, Pole Angle: {pole_angle_deg:.2f}°")

        total_reward += reward

    print(f"Episode {episode + 1}: Total Reward = {total_reward}")

env.close()  # Close the environment
```
